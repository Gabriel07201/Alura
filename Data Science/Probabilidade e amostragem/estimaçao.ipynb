# Nível de confiança e significância

## Intervalo de segurança

### Exemplo

Desvio populacional de 150. Amostra aleatória de 20 sacos de um lote específico, obteve um peso médio de 5050. Nível de significância de 5%

```python
media_amostral = 5050
significancia = 0.05
```

```python
confianca = 1 - significancia
confianca
```

obtendo z

```python
0.95 / 2
```

```python
0.5 + (0.95 / 2)
```

```python
z = 1.96
```

```python
from scipy.stats import norm
z = norm.ppf(0.975)
z
```

Obtendo o sigma sobre raiz de n

```python
desvio_padrao = 150
n = 20
```

```python
raiz_n = np.sqrt(n)
raiz_n
```

```python
sigma = desvio_padrao / raiz_n
sigma
```

obtendo e 

```python
e = z * sigma
e
```

Solução 1 para o intervalo de segurança

```python
intervalo = (
    media_amostral - e,
    media_amostral + e
)
intervalo
```

Solução 2 para o invervalo de segurança

```python
norm.interval(alpha = 0.95, loc = media_amostral, scale = sigma)
```
